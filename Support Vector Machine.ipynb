```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("letterdata.csv")
```

```python
data
```

```python
(data.describe)
```

```python
x = data.drop("letter" , axis = 1)
```

```python
y = data["letter"] 
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
# x = x_train , x_test
# y = y_train , y_test
# test_size = y = 20% of dataset
#
x_train,x_test,y_train,y_test = train_test_split(x , y, test_size = 0.20)
```

```python
#creating object as svclassifier for support vector classifier
svclassifier = SVC(kernel = "linear")
#fitting the model by .fit
svclassifier.fit(x_train,y_train)
```

```python
#predicting the test set results
y_predict = svclassifier.predict(x_test)
```

```python
print(y_predict)
print(y_test)
```

```python
#accuracy of the algorithm on the dataset
print (svclassifier.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
# confusion matrix
mat = confusion_matrix(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
